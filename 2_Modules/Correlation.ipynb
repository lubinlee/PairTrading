```python
def corr_pair(dataframe,absolute = False):
    if absolute == False:
        stdCor = 'cor'
    else:
        stdCor = 'absCor'
    # 상관관계 dataframe 으로 변환
    dataframe = dataframe.corr()
    # pair1, pair2, correlation값으로 이뤄진 시리즈(Series) 만들기
    dataframe = dataframe.unstack()
    # 중복 제거
    dataframe = dataframe.drop_duplicates().dropna()
    # Series -> Dataframe
    dataframe = dataframe.to_frame()
    # 절대값이 높은 순으로 정렬
    dataframe['absCor'] = dataframe[0].abs()
    dataframe = dataframe.reset_index()
    dataframe.columns=['pair1','pair2','cor','absCor']
    dataframe = dataframe.sort_values(by=[stdCor],ascending=[False])
    dataframe = dataframe.drop(0,0)
    dataframe = dataframe.reset_index(drop=True)
    # column명 변경
    #dataframe.columns=['pair1','pair2','cor','absCor']
    return dataframe
```

```python
# 상관관계 Pair 스크리닝 함수(filtering) 
def sub_pair(dataframe,value=None, ratio=None, down = False, absolute = False):
    if absolute == False:
        stdCor = 'cor'
        stdValue = value
    else:
        stdCor = 'absCor'
        if value is not None:
            stdValue = abs(value)
        
    ## 상위 부분집합
    if down == False:
        # value와 ratio 모두 입력되지 않으면, 상관계수가 평균이상인 부분집합
        if (value is None) and (ratio is None):
            averCorr = dataframe[stdCor].mean()
            outputDf = dataframe[dataframe[stdCor]>=averCorr]
            return outputDf
        # value만 입력되었을 때, 상관계수가 입력된 value값 이상인 부분집합
        elif (value is not None) and (ratio is None):
            if value <=1 and value >= -1:
                outputDf = dataframe[dataframe[stdCor]>= stdValue]
                return outputDf
            else:
                print('value는 -1과 1사이의 수를 입력하여 주세요')
        # ratio만 입력되었을 때, 전체 상관계수 쌍에서 상위 (입력된 ratio값) 부분집합
        elif (value is None) and (ratio is not None):
            if ratio <= 1 and ratio >= 0:
                ratioIndex = int(len(dataframe)*ratio)
                outputDf = dataframe.head(ratioIndex)
                return outputDf
            else:
                print('ratio는 0과 1사이의 수를 입력하여 주세요.')
        # value 와 ratio 모두 입력되었을 때는 더 상위 부분집합 
        else:
            if ratio <= 1 and ratio >= 0 and value <=1 and value >=-1:
                ratio_index = int(len(dataframe)*ratio)      
                if value >= dataframe.iloc[ratioIndex,3]:
                    outputDf = dataframe[dataframe[stdCor]> stdValue]
                    return outputDf
                else:
                    outputDf = dataframe.head(ratioIndex)
                    return outputDf
            else:
                print('value는 -1과 1사이, ratio는 0과 1사이의 수를 입력하여 주세요.')
    ## 하위 부분집합
    else:
         # value와 ratio 모두 입력되지 않으면, 상관계수가 평균이하인 부분집합
        if (value is None) and (ratio is None):
            averCorr = dataframe[stdCor].mean()
            outputDf = dataframe[dataframe[stdCor]<=averCorr]
            outputDf = outputDf.sort_values(by=[stdCor],ascending=[True])
            outputDf = outputDf.reset_index(drop=True)
            return outputDf
        # value만 입력되었을 때, 상관계수가 입력된 value값 이하인 부분집합
        elif (value is not None) and (ratio is None):
            if value <=1 and value >= -1:
                outputDf = dataframe[dataframe[stdCor]<= stdValue]
                outputDf = outputDf.sort_values(by=[stdCor],ascending=[True])
                outputDf = outputDf.reset_index(drop=True)
                return outputDf
            else:
                print('value는 -1과 1사이의 수를 입력하여 주세요')
        # ratio만 입력되었을 때, 전체 상관계수 쌍에서 하위 (입력된 ratio값) 부분집합
        elif (value is None) and (ratio is not None):
            if ratio <= 1 and ratio >= 0:
                ratioIndex = int(len(dataframe)*ratio)
                outputDf = dataframe.tail(ratioIndex)
                outputDf = outputDf.sort_values(by=[stdCor],ascending=[True])
                outputDf = outputDf.reset_index(drop=True)
                return outputDf
            else:
                print('ratio는 0과 1사이의 수를 입력하여 주세요.')
        # value 와 ratio 모두 입력되었을 때는 더 하위 부분집합 
        else:
            if ratio <= 1 and ratio >= 0 and value <=1 and value >=-1:
                ratio_index = int(len(dataframe)*ratio)      
                if value <= dataframe.iloc[ratioIndex,3]:
                    outputDf = dataframe[dataframe[stdCor]< stdValue]
                    outputDf = outputDf.sort_values(by=[stdCor],ascending=[True])
                    outputDf = outputDf.reset_index(drop=True)
                    return outputDf
                else:
                    outputDf = dataframe.tail(ratioIndex)
                    outputDf = outputDf.sort_values(by=[stdCor],ascending=[True])
                    outputDf = outputDf.reset_index(drop=True)
                    return outputDf
            else:
                print('value는 -1과 1사이, ratio는 0과 1사이의 수를 입력하여 주세요.')
```
